```python
df = pd.read_csv("world_population (2).csv")
```

```python
top10 = df.head(10)
plt.figure(figsize=(10,5))
plt.bar(
    top10['Location'],
    top10['Population'])
plt.xticks(rotation=45)
plt.title("Top 10 Most Populated Countries")
plt.show()
```

**India and China dominate the global population landscape. Together they account for a significant share of the world's population.**

```python
plt.figure(figsize=(8,5))
plt.hist(df['Population'])
plt.title("Population Distribution")
plt.show()
```

**Most countries have relatively small populations, while only a few countries have extremely large populations.**

```python
top15 = df.head(15)
plt.figure(figsize=(10,5))
plt.barh(
    top15['Location'],
    top15['Population'])
plt.title("Top 15 Countries")
plt.show()
```

```python
top5 = df.head(5)
plt.figure(figsize=(8,8))
plt.pie(
    top5['Population'],
    labels=top5['Location'],
    autopct='%1.1f%%')
plt.title("Top 5 Population Share")
plt.show()
```

**The top countries contribute a major share of the total population.**


```python
sns.boxplot(x=df['Population'])
plt.title("Population Outliers")
plt.show()
```

**Several countries appear as population outliers.**


Global Population Distribution

The analysis reveals that population is concentrated in a small number of countries, with India and China leading significantly. Most countries have populations below the global average, indicating a highly skewed distribution.

The visualizations highlight clear population inequalities across nations. These insights can help governments, policymakers, and businesses make informed decisions regarding resource allocation, infrastructure planning, healthcare services, and market expansion strategies.

Key Findings:
1. Population is concentrated in a few countries.
2. Most countries have populations below the average.
3. Significant outliers exist.
4. Population distribution is positively skewed.

Conclusion:
Understanding population distribution is essential for effective planning and decision-making at both national and international levels.
